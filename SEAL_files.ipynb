```python
# Donwload dataset
import os
import json
import random

# Paths
CONVO_PATH = "few-shot/ThousandVoicesOfTrauma/conversations"
META_PATH = "few-shot/ThousandVoicesOfTrauma/metadata"


# Where to save SEAL files
OUTPUT_PATH = "my_dataset"
os.makedirs(OUTPUT_PATH, exist_ok=True)
```

```python
conversation_files = sorted(os.listdir(CONVO_PATH))
metadata_files = sorted(os.listdir(META_PATH))

len(conversation_files), len(metadata_files)
```

```python
# Build SEAL Challenge + Solution Files From Your Dataset
import json
import os

all_challenges = []
all_solutions = []

for convo_file in os.listdir(CONVO_PATH):
    if not convo_file.endswith("_conversation.json"):
        continue

    # Load conversation
    with open(os.path.join(CONVO_PATH, convo_file), "r") as f:
        convo_data = json.load(f)

    # full conversation text
    full_text = " ".join(convo_data.get("full_conversation", []))

    # derive metadata filename
    meta_file = convo_file.replace("_conversation.json", "_metadata.json")
    meta_path = os.path.join(META_PATH, meta_file)

    if not os.path.exists(meta_path):
        # skip if metadata is missing
        continue

    # Load metadata
    with open(meta_path, "r") as f:
        meta = json.load(f)

    true_cues = meta["client_profile"].get("exhibited_behaviors", [])

    # Add challenge entry
    all_challenges.append({
        "id": convo_file,
        "input_text": full_text,
        "model_predicted_cues": [], 
        "task_type": "cue_detection"
    })

    # Add solution entry
    all_solutions.append({
        "id": convo_file,
        "correct_output": {
            "true_cues": true_cues
        }
    })

# Save files
with open(os.path.join(OUTPUT_PATH, "my_challenges.json"), "w") as f:
    json.dump(all_challenges, f, indent=2)

with open(os.path.join(OUTPUT_PATH, "my_solutions.json"), "w") as f:
    json.dump(all_solutions, f, indent=2)

len(all_challenges), len(all_solutions)
```
